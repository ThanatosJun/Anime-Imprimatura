```python
import pandas as pd
import cv2


def get_color_name(r, g, b):
    min_diff = 10000
    color_name = ''
    for i in range(len(csv_df)):
        d = abs(r - int(csv_df.loc[i, "R"])) + abs(g - int(csv_df.loc[i, "G"]))+ abs(b - int(csv_df.loc[i, "B"]))
        if d <= min_diff:
            min_diff = d
            color_name = csv_df.loc[i,"color_name"]
    return color_name


def click_info(event, x, y, flags, param):
    # 只处理双击事件
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b, g, r, xpos, ypos, clicked
        xpos = x
        ypos = y
        b, g, r = img[y, x]     # 获取b, g, r
        b = int(b)
        g = int(g)
        r = int(r)
        clicked = True


r = g = b = xpos = ypos = 0
clicked = False
img = cv2.imread('./Camera_XHS_171008966829503029o01kfp5ihwuzr5010y4w099as2xz5.jpg')
index = ["color", "color_name", "hex", "R", "G", "B"]
csv_df = pd.read_csv('color.csv', names=index, header=None)

cv2.namedWindow('image')
cv2.setMouseCallback('image',  click_info)

while True:
    output_img = img.copy()  # Create a copy of the image for display

    if clicked:
        # Draw a circle to mark the selected point
        cv2.circle(output_img, (xpos, ypos), 5, (255, 0, 0), -1)

        # Draw text displaying the coordinates of the selected point
        text = f'({xpos}, {ypos})'
        cv2.putText(output_img, text, (xpos + 10, ypos - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)

        # Draw rectangle and display color information as before
        cv2.rectangle(output_img, (20, 20), (750, 60), (b, g, r), -1)
        color_text = get_color_name(r, g, b) + ' R=' + str(r) + ' G=' + str(g) + ' B=' + str(b)
        cv2.putText(output_img, color_text, (50, 50), 2, 0.8, (255, 255, 255), 2, cv2.LINE_AA)
        if (r + g + b >= 600):
            cv2.putText(output_img, color_text, (50, 50), 2, 0.8, (0, 0, 0), 2, cv2.LINE_AA)

    cv2.imshow("image", output_img)
    if clicked:
        clicked = False

    # 点击 esc 键
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
